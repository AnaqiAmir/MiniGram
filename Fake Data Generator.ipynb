# Fake Data Generator

## Setup

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from faker import Faker
import random

# Initialize Faker for generating realistic usernames
faker = Faker()

# Extract the range of dates from the sample data for the 'created_at' attribute
date_range_start = pd.to_datetime("2014-01-01 00:00:00")
date_range_end = pd.to_datetime("2024-12-31 23:59:59")
```

## Users

```python
# Function to generate synthetic usernames based on patterns from the sample data
def generate_username():
    return faker.user_name()

# Function to generate synthetic timestamps within the desired range
def generate_user_timestamp():
    return faker.date_time_between(start_date=date_range_start, end_date=date_range_end).strftime('%Y-%m-%d %H:%M:%S')

# Generate synthetic data
synthetic_user_data = pd.DataFrame({
    'id': range(1, 10001),
    'username': [generate_username() for _ in range(10000)],
    'created_at': [generate_user_timestamp() for _ in range(10000)]
})

# Display the first few rows of the synthetic data
synthetic_user_data.head()
```

# Photos

```python
# Function to generate synthetic image urls
def generate_image_url():
    return faker.image_url()

# Function to generate user id based on user column
def generate_user_id():
    return random.randint(synthetic_user_data['id'].min(), synthetic_user_data['id'].max())

# Function to generate synthetic timestamps within the desired range
def generate_image_timestamp(user_created_at):
    return faker.date_time_between(start_date=user_created_at, 
                                   end_date=date_range_end).strftime('%Y-%m-%d %H:%M:%S')

# Generate synthetic data
synthetic_image_data = pd.DataFrame({
    'id': range(1, 75001),
    'image_url': [generate_image_url() for _ in range(75000)],
    'user_id': [generate_user_id() for _ in range(75000)],
    'created_at': [None] * 75000 # Placeholder for timestamps
})

# Generating timestamps where the start time has to be after the user has created their profile
for index, row in synthetic_image_data.iterrows():
    user_created_at = synthetic_user_data.loc[synthetic_user_data['id'] == row['user_id'], 'created_at'].values[0]
    user_created_at = pd.to_datetime(user_created_at)
    synthetic_image_data.at[index, 'created_at'] = generate_image_timestamp(user_created_at)

# Display data
synthetic_image_data.head()
```
